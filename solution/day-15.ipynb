# Crosstab

```python
#crosstab is used to when we want to check the relationship between two or more attributes
pd.crosstab(customer['Education'],customer['Income'])
```

```python
#transpose crosstab
pd.crosstab(customer['Education'],customer['Marital_Status']).T
```

```python
#cross tab to get percentage
pd.crosstab(customer['Education'],customer['Marital_Status']).apply(lambda r: (r/len(customer))*100, axis=1)
```

# Pivot Table

```python
pd.pivot_table(customer, index=['Education','Marital_Status'], values="Income", aggfunc="mean")
```

Difference bewteen Pivot and Crosstab
1. Pivot accepts data in dataframe format only which is not the case with Crosstab.
2. Crosstab source code contains the instruction to convert data into dataframe and then call Pivot Table function for futher execution.
3. Or we should use Pivot Table if data is in dataframe format
