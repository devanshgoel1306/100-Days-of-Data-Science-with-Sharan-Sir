```python
#creating a dataframe using pandas
heart= pd.read_csv('../input/heart-failure-prediction/heart.csv')
```

```python
heart.columns
```

```python
#1. rename a column of a dataframe
heart1= heart.rename(columns= {'Sex':'Gender'})
print(heart1.columns)
```

```python
#2. rename index
#take only 26 records from dataframe
heart2= heart[:26]
print('Old dataframe')
print(heart2.head())

new_index= dict()
for i in range(26):
    new_index[i]= chr(97+i)
    
#print(new_index)
heart2= heart2.rename(index= new_index)
print('New dataframe')
print(heart2.head())
```

```python
#3. give name to axis
heart3= heart.rename_axis('Index',axis='rows').rename_axis('Attributes',axis='columns')
print(heart3.head())
print(heart3.columns)
```

```python
#4. giving new values to each column
new_column= list()

for column in heart.columns:
    new_column.append(column.lower())
    
#print(new_column)
from copy import deepcopy
heart4= deepcopy(heart)
heart4.columns= new_column

print(heart4.head())
```

```python
#5. replace a default string in header
heart5= deepcopy(heart)
heart5.columns= heart5.columns.str.replace('_','__')
print(heart5.head())
```

```python
#6. replacing value of a particular column using loc
#2 is the index
print(heart.loc[2,'MaxHR'])

heart.loc[2,'MaxHR']= 100
print(heart.loc[2,'MaxHR'])
```

```python
#7. replacing value of a particular column using loc
#2 is the index
print(heart.loc[4,'MaxHR'])

heart.loc[4,'MaxHR']= 120
print(heart.loc[4,'MaxHR'])
```

```python
#8. replacing error codes
print(heart.head())
heart.replace(172,170,inplace=True)
print(heart.head())
```

```python
#9. replace a list of items
print(heart.replace([0,1],[1,0]))
```
