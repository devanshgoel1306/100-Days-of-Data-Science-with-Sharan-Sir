**Combine, Merge and Join****

```python
#import the dataset
corona= pd.read_csv('../input/corona-virus-report/country_wise_latest.csv')
```

```python
#print(corona.head())
display(corona[0:10])
```

```python
#1. Create two dataframes out of corona of size 5 each and datasets must overlap
corona1= corona[:5]
corona2= corona[3:8]
```

```python
display(corona1.tail())
```

```python
display(corona2.tail())
```

```python
#2. use concat to concatenate two dataframes
#concat always uses outer join
pd.concat([corona1,corona2],axis=0)
```

```python
pd.concat([corona1,corona2],axis=1)
```

```python
#3. use merge for inner, outer, left and right join
#there is no axis parameter in merge function
#index will get automatically reset
from copy import deepcopy
```

```python
corona3= corona1.merge(corona2, how= 'outer')
display(corona3)
```

```python
print(corona3.shape)
print(corona1.shape)
print(corona2.shape)
```

```python
corona4= corona1.merge(corona2, how= 'inner')
display(corona4)
```

```python
display(corona4.shape)
```

```python
corona5= corona1.merge(corona2, how= 'left')
display(corona5)
```

```python
print(corona5.shape)
```

```python
corona6= corona1.merge(corona2, how= 'right')
display(corona6)
```

```python
print(corona6.shape)
```

```python
#right join can be implemented using left join also
#just interchange name of dataframe while calling merge
corona7= corona2.merge(corona1, how= 'right')
display(corona7)
```

```python
#4. use join method
#join method uses left join by default
#left and right suffix are used to differentiate the columns that are present in both the dataframes
new_df= corona1.join(corona2,lsuffix='_left',rsuffix='_right')
```

```python
display(new_df)
```

```python
#using join and a common column as a key to make the join
display(corona1.set_index('Country/Region').join(corona2.set_index('Country/Region'),lsuffix='_left',rsuffix='_right'))
```

```python
#to preserve the index in the original dataframe then we use below method
display(corona1.join(corona2.set_index('Country/Region'),on='Country/Region',lsuffix='_left',rsuffix='_right'))
```
